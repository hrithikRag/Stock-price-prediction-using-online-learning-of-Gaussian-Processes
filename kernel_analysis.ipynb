```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Kernel initialization

```python
def RBF(x,y,length_scale=10):
    return np.exp((x-y)**2/(-length_scale))
```

```python
import numpy as np

# Define the Matern kernel function
def Mt(x, y, length_scale, nu):
    dist = np.linalg.norm(x - y)
    if nu == 0.5:
        return np.exp(-dist / length_scale)
    elif nu == 1.5:
        return (1.0 + np.sqrt(3.0) * dist / length_scale) * np.exp(-np.sqrt(3.0) * dist / length_scale)
    elif nu == 2.5:
        return (1.0 + np.sqrt(5.0) * dist / length_scale + 5.0 * dist**2 / (3.0 * length_scale**2)) \
               * np.exp(-np.sqrt(5.0) * dist / length_scale)
```

```python
def ExpSS(x, y, length_scale, periodicity):
    dist = np.linalg.norm(x - y)
    sin_term = np.sin(np.pi * dist / periodicity)
    return np.exp(-2.0 * sin_term**2 / length_scale**2)
```

```python
def C(c):
    return c
```

# Kernel Combinations

```python
def kernel_1(x,y):
    return C(12.8**2)*Mt(x,y,length_scale=2.08, nu=1.5)+ExpSS(x,y,length_scale=3.08, periodicity=6)
```

```python
def kernel_2(x,y):
    return C(12.8**2)*Mt(x,y,length_scale=2.08, nu=2.5)
```

```python
def kernel_3(x,y):
    return C(12.8**2)*RBF(x,y,length_scale=20)
```

```python
def kernel_4(x,y):
    return C(12.8**2)*ExpSS(x,y,length_scale=2.08, periodicity=6)+C(12.8**2)*RBF(x,y,length_scale=5)
```

```python
def kernel_5(x,y):
    return C(12.8**2)*ExpSS(x,y,length_scale=2.08, periodicity=6)+C(5**2)*RBF(x,y,length_scale=5)
```

```python
def kernel_6(x,y):
    return C(12.8**2)*ExpSS(x,y,length_scale=2.08, periodicity=6)+RBF(x,y,length_scale=5)
```

```python
def kernel_7(x,y):
    return C(12.8**2)*Mt(x,y,length_scale=2.08, nu=1.5)+RBF(x,y,length_scale=5)
```

```python
def kernel_8(x,y):
    return Mt(x,y,length_scale=2.08, nu=2.5)+C(12.8**2)*RBF(x,y,length_scale=5)
```

# Kernel Analysis

```python
y_arr=np.linspace(-10,10,63)
print(len(y_arr))
```

```python
K1_arr=[]
K2_arr=[]
K3_arr=[]
K4_arr=[]
K5_arr=[]
K6_arr=[]
K7_arr=[]
K8_arr=[]
x=np.array([0])
for i in range(len(y_arr)):
    y=np.array(y_arr[i])
    k1=kernel_1(x,y)
    k2=kernel_2(x,y)
    k3=kernel_3(x,y)
    k4=kernel_4(x,y)
    k5=kernel_5(x,y)
    k6=kernel_6(x,y)
    k7=kernel_7(x,y)
    k8=kernel_8(x,y)
    K1_arr.append(k1)
    K2_arr.append(k2)
    K3_arr.append(k3)
    K4_arr.append(k4)
    K5_arr.append(k5)
    K6_arr.append(k6)
    K7_arr.append(k7)
    K8_arr.append(k8)
```
